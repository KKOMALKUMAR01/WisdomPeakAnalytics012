```python
import os, random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    roc_curve
)
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
```

```python
import kagglehub

path = kagglehub.dataset_download("paultimothymooney/chest-xray-pneumonia")
print("Downloaded path:", path)

def find_chest_xray_root(download_path):
    for root, dirs, files in os.walk(download_path):
        if "chest_xray" in dirs:
            return os.path.join(root, "chest_xray")
    return None

chest_root = find_chest_xray_root(path)
print("chest_root:", chest_root)
print("chest_root contents:", os.listdir(chest_root))

train_path = os.path.join(chest_root, "train")
val_path   = os.path.join(chest_root, "val")
test_path  = os.path.join(chest_root, "test")

print("Train:", os.listdir(train_path))
print("Val:", os.listdir(val_path))
print("Test:", os.listdir(test_path))
```

```python
#dataprocessing
# Raw (for visualization)
raw_tfm = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Preprocess (for feature extraction)
preprocess_tfm = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# Augmentation (for report figure)
augment_tfm = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=1.0),
    transforms.RandomRotation(degrees=10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor()
])
```

```python
train_ds = datasets.ImageFolder(train_path, transform=preprocess_tfm)
val_ds   = datasets.ImageFolder(val_path,   transform=preprocess_tfm)
test_ds  = datasets.ImageFolder(test_path,  transform=preprocess_tfm)

raw_ds = datasets.ImageFolder(train_path, transform=raw_tfm)
aug_ds = datasets.ImageFolder(train_path, transform=augment_tfm)

class_names = train_ds.classes
print("Classes:", class_names)
print("Train size:", len(train_ds), "Val size:", len(val_ds), "Test size:", len(test_ds))
```

```python
def make_subset(ds, n):
    idx = np.random.choice(len(ds), size=min(n, len(ds)), replace=False)
    return Subset(ds, idx)

#  Start small; if everything is fine, increase these numbers
train_small = make_subset(train_ds, 2000)
test_small  = make_subset(test_ds,  800)

train_loader = DataLoader(train_small, batch_size=16, shuffle=False, num_workers=0)
test_loader  = DataLoader(test_small,  batch_size=16, shuffle=False, num_workers=0)

print("Subset sizes:", len(train_small), len(test_small))
```

```python
resnet = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
resnet.fc = nn.Identity()
resnet = resnet.to(device)
resnet.eval()
```

```python
def extract_features(loader):
    feats, labels = [], []
    with torch.no_grad():
        for i, (x, y) in enumerate(loader):
            x = x.to(device)
            f = resnet(x)               # (batch, 512)
            feats.append(f.numpy())
            labels.append(y.numpy())
            if i % 10 == 0:
                print(f"Processed batch {i}/{len(loader)}")
    return np.vstack(feats), np.concatenate(labels)

X_train, y_train = extract_features(train_loader)
X_test, y_test   = extract_features(test_loader)

print("Feature shapes:", X_train.shape, X_test.shape)
```

```python
SEED = 42 # Define SEED for reproducibility
X_std = StandardScaler().fit_transform(X_train)

pca = PCA(n_components=2, random_state=SEED)
X_pca = pca.fit_transform(X_std)

print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total explained variance:", pca.explained_variance_ratio_.sum())

plt.figure(figsize=(6,5))
for cls in np.unique(y_train):
    mask = (y_train == cls)
    plt.scatter(X_pca[mask, 0], X_pca[mask, 1],
                label=class_names[int(cls)], s=10, alpha=0.7)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA-based 2D Projection of CNN Features")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "pca_scatter.png"), dpi=300)
plt.show()
```

```python
def compute_binary_metrics(y_true, y_pred, y_prob):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)          # Positive class = 1 (PNEUMONIA)
    sens = recall_score(y_true, y_pred)             # Sensitivity/Recall
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_prob)

    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    spec = tn / (tn + fp) if (tn + fp) > 0 else 0.0
    return acc, prec, sens, spec, f1, auc
```

```python
models_dict = {
    "Logistic Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(max_iter=3000))
    ]),
    "LDA": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LinearDiscriminantAnalysis())
    ]),
    "Naive Bayes": GaussianNB(),
    "SVM (RBF)": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", SVC(kernel="rbf", probability=True))
    ]),
    "Gradient Boosting": GradientBoostingClassifier(random_state=SEED)
}
```

```python
rows = []
saved_preds = {}  # store predictions for later confusion matrix / ROC

for name, model in models_dict.items():
    print("\nTraining:", name)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        # fallback normalization if needed
        scores = model.decision_function(X_test)
        y_prob = (scores - scores.min()) / (scores.max() - scores.min() + 1e-9)

    acc, prec, sens, spec, f1, auc = compute_binary_metrics(y_test, y_pred, y_prob)

    rows.append({
        "Model": name,
        "Accuracy": acc,
        "Precision": prec,
        "Sensitivity (Recall)": sens,
        "Specificity": spec,
        "F1-score": f1,
        "ROC-AUC": auc
    })

    saved_preds[name] = (y_pred, y_prob)

table_iii = pd.DataFrame(rows).sort_values("ROC-AUC", ascending=False)
table_iii
```

```python
best_model_name = table_iii.iloc[0]["Model"]
print("Best model:", best_model_name)

y_pred_best, y_prob_best = saved_preds[best_model_name]
```

```python
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(5,4))
plt.imshow(cm)
plt.colorbar()
plt.title(f"Confusion Matrix  {best_model_name}")
ticks = np.arange(len(class_names))
plt.xticks(ticks, class_names, rotation=45)
plt.yticks(ticks, class_names)

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_best, target_names=class_names))
```

```python
fpr, tpr, _ = roc_curve(y_test, y_prob_best)
auc = roc_auc_score(y_test, y_prob_best)

plt.figure(figsize=(5,4))
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0,1],[0,1],'--', label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve  {best_model_name}")
plt.legend()
plt.tight_layout()
plt.show()
```
